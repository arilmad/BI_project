### Create project database

```python
conn = mysql.connector.connect(user=d['user'], password=d['pw'])
cur = conn.cursor()

DB_NAME = 'bi_project'
```

```python
def create_database(cursor):
    try:
        cursor.execute(
            "CREATE DATABASE {} DEFAULT CHARACTER SET 'utf8'".format(DB_NAME))
    except mysql.connector.Error as err:
        print("Failed creating database: {}".format(err))
        exit(1)
try:
    cur.execute("USE {}".format(DB_NAME))
except mysql.connector.Error as err:
    print("Database {} does not exists.".format(DB_NAME))
    if err.errno == errorcode.ER_BAD_DB_ERROR:
        create_database(cur)
        print("Database {} created successfully.".format(DB_NAME))
        conn.database = DB_NAME
    else:
        print(err)
        exit(1)
```

#### Alternative 2: Preprocess `.json` files to `.csv` in `frost_api_to_json.ipynb` and load `.csv` files into weather table

```python
query = """
    drop table if exists Weather_Gathered
"""
cur.execute(query)

query = """
    create temporary table Weather_Gathered (
        a int,
        b int, 
        element_id varchar(30),
        c int,
        level varchar(70),
        perf varchar(2),
        d int, 
        timeoff varchar(10),
        timeres varchar(10),
        times_id varchar(2),
        unit varchar(5),
        value varchar(10),
        referencetime datetime not null,
        source_id varchar(20) not null,
        primary key(referencetime, source_id)
    )
"""
cur.execute(query)
```

```python
folder = f'{cwd}//src//weather_csvs//'

for file in os.listdir(folder):
    if file.split('.')[-1] != 'csv': continue
    file_path = os.path.join(folder, file)
    
    query = f"""
        load data infile '{file_path}'
        into table Weather_Gathered
        fields terminated by ',' 
        enclosed by '"'
        lines terminated by '\r\n'
        ignore 1 rows  
    """
    cur.execute(query)
```

```python
query = """
    select *
    from Weather_Gathered
    limit 3
"""
pd.read_sql(query, conn)
```

```python
query = """
    drop table if exists Dim_Weather
"""
cur.execute(query)

query = """
    create table Dim_Weather as (
        select 
            referencetime as time,
            trim(trailing ':0' from source_id) as weather_station,
            value as temperature
        from Weather_Gathered
    )
"""
cur.execute(query)

query = """
    alter table Dim_Weather
    add primary key(time, weather_station)
"""
cur.execute(query)
```

```python
query = """
    select 
        weather_station,
        count(*)
    from Dim_Weather
    group by weather_station
"""
pd.read_sql(query, conn).T
```

```python
query = """
    drop table if exists Facts
"""
cur.execute(query)

query = """
    create table Facts
    as(
        select 
            bel.time as time,
            date(bel.time) as day,
            bel.loc_id,
            s.nearest_station as weather_station,
            bel.consumption_kvah
        from Belastning_Gathered bel 
        left join
        (
            select 
                m.loc_id as loc_id,
                mns.nearest_station
            from 
                Dim_Municipality m,
                Municipality_Nearest_Station mns
            where m.municipality = mns.municipality
        ) s
        on bel.loc_id = s.loc_id  
    )
"""
cur.execute(query) 
```

```python
query = """
    select *
    from Facts
    where loc_id = 3
    limit 3
"""
pd.read_sql(query, conn)
```

```python
query = """
    select count(*)
    from Facts
"""
pd.read_sql(query, conn)
```

## Sample queries

i.  Find the top five locations with the highest average consumption and their municipality.

```python
query = """
    with a as (
        select 
            f.loc_id,
            avg(consumption_kvah) as avg_con,
            municipality,
            rank() over (
                order by avg(consumption_kvah) desc
            ) con_rank
            
        from Facts f
            left join Dim_Municipality m on f.loc_id = m.loc_id
        group by f.loc_id
    )
    select
        loc_id,
        municipality,
        avg_con,
        con_rank
        
        from a
        
        having con_rank <= 5
        
"""
pd.read_sql(query, conn)
```

ii. For those five, what was the mean temperature, oil price and el spot price at their day in 2019 of maximum consumption?

```python
query = """
    with a as (
        select 
            f.loc_id,
            avg(consumption_kvah) as avg_con,
            avg(temperature) as avg_temp,
            avg(o.price) as avg_oilspot,
            avg(e.price) as avg_elspot,
            f.day,
            rank() over (
                partition by f.loc_id
                order by avg(consumption_kvah) desc
            ) con_rank 
            
        from Facts f
            left join Dim_Weather w on (w.time = f.time and w.weather_station = f.weather_station)
            left join Dim_Oilspot o on o.day = f.day
            left join Dim_Elspot e on e.time = f.time
            
        where 
            extract(year from f.time) = 2019 and
            f.loc_id in (1, 7, 9, 12 ,13)
        
        group by 
            f.day,
            f.loc_id
    )
    
    select * 
    from a
    having con_rank = 1  
    limit 10
    
        
"""
pd.read_sql(query, conn)
```

Did anyone exceed the tolerance of their transformer?

```python
query = """
    select 
        f.loc_id,
        time,
        consumption_kvah,
        upper_tol_kvah
        
    from 
        Facts f left join Dim_Transformer t on f.loc_id = t.loc_id
    where cast(consumption_kvah as decimal) > cast(upper_tol_kvah as decimal)
    limit 20
"""
pd.read_sql(query, conn)
```

Apparently loc_id 2 did on Christmas Eve 2018 and later in January!

### Join all interesting data to output table

```python
query = """
    select
        f.time as time,
        f.loc_id as location_id,
        consumption_kvah as el_consumption,
        w.temperature as degC,
        o.price as oil_price,
        e.price as el_price,
        usd.price_1_USD as price_1_USD
        
    from Facts f
    left join Dim_Weather w on (f.time=w.time and f.weather_station=w.weather_station) 
    left join Dim_Oilspot o on (f.day=o.day)
    left join Dim_Elspot e on (f.time=e.time)
    left join Dim_USD_NOK usd on (f.day = usd.day)
"""
df = pd.read_sql(query, conn)
```
